```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 5, 7, 10, 15])

# Choose polynomial degree
degree = 2

# Construct design matrix
X = np.vstack([x**i for i in range(degree + 1)]).T

# Compute coefficients using Ordinary Least Squares
beta = np.linalg.inv(X.T @ X) @ (X.T @ y)

# Predict outputs for original x
y_pred = X @ beta

print("Polynomial Coefficients:", beta)

# For smooth curve
x_smooth = np.linspace(1, 5, 200)
X_smooth = np.vstack([x_smooth**i for i in range(degree + 1)]).T
y_smooth_pred = X_smooth @ beta

# Visualization
plt.figure(figsize=(8,5))
plt.scatter(x, y, color='red', label='Data Points', s=50)
plt.plot(x_smooth, y_smooth_pred, color='blue', label='Polynomial Fit (Degree 2)')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Polynomial Regression Example")
plt.legend()
plt.grid(True)
plt.show()
```
